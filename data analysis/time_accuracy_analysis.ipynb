```python
import matplotlib.pyplot as plt

# Data
res_labels = ["32x32", "64x64", "128x128", "256x256", "512x512"]
time_minutes = [0.82, 1.22, 5.20, 11.62, 43.24]
accuracy = [0.68, 0.73, 0.78, 0.72, 0.72]

plt.figure(figsize=(10, 6))

# Scatter plot only
plt.scatter(time_minutes, accuracy, s=120, color="#1f77b4", edgecolor="black", zorder=3)

# Label each point with resolution
for i, label in enumerate(res_labels):
    plt.text(time_minutes[i] + 0.5, accuracy[i] + 0.005, label,
             fontsize=11, weight="bold", color="#333333")

# Titles and labels
plt.title("Trade-off: Training Time vs Accuracy", fontsize=18, weight="bold")
plt.xlabel("Training Time (minutes)", fontsize=12)
plt.ylabel("Accuracy", fontsize=12)

# Axes limits
plt.xlim(0, max(time_minutes) + 10)
plt.ylim(0.6, 0.85)

# Subtle grid
plt.grid(True, linestyle="--", alpha=0.3, zorder=1)

# Ticks styling
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)

plt.tight_layout()
plt.savefig("training_vs_accuracy.png", dpi=300, bbox_inches="tight")
plt.show()
```
