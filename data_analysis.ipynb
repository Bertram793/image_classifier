```python
import matplotlib.pyplot as plt

# Data
res_labels = ["32×32", "64×64", "128×128", "256×256", "512×512"]
accuracy = [0.71, 0.76, 0.80, 0.74, 0.69]

# Even spacing positions
x = range(len(res_labels))

plt.figure(figsize=(10, 6))
plt.plot(x, accuracy, marker="o", linewidth=2)

# Labels on points
for i, y in enumerate(accuracy):
    plt.text(i, y + 0.015, f"{y*100:.1f}%", ha="center", fontsize=11)

plt.title("Resolution vs Accuracy", fontsize=18, weight="bold")
plt.xlabel("Image Resolution (pixels × pixels)", fontsize=12)
plt.ylabel("Accuracy", fontsize=12)

plt.xticks(x, res_labels)
plt.ylim(0, 1.05)
plt.grid(True, linestyle="--", alpha=0.4)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Example data
resolutions = [32, 64, 128, 256, 512]
accuracy = [0.71, 0.76, 0.80, 0.74, 0.69]
train_time = [2.06, 2.68, 5.59, 14.55, 52.95]
fig, ax1 = plt.subplots(figsize=(8,5))

# Accuracy line (left axis)
ax1.plot(resolutions, accuracy, marker="o", linewidth=2)
ax1.set_xlabel("Image resolution (px)")
ax1.set_ylabel("Accuracy")
ax1.set_ylim(0, 1)

# Time line (right axis)
ax2 = ax1.twinx()
ax2.plot(resolutions, train_time, marker="s", linewidth=2)
ax2.set_ylabel("Training time (minutes)")

plt.title("Resolution vs Accuracy and Training Time")
plt.grid(True)
plt.show()
```
